### 1. Supervised Learning
This involves training models on labeled data (input-output pairs) to make predictions.

*   **Classification**: Predicting a categorical output (e.g., spam/not spam, disease/no disease).
    *   *Scikit-learn examples*: `LogisticRegression`, `SVC`, `RandomForestClassifier`.
*   **Regression**: Predicting a continuous numerical output (e.g., house prices, temperature).
    *   *Scikit-learn examples*: `LinearRegression`, `SVR`, `RandomForestRegressor`.

### 2. Unsupervised Learning
This involves finding patterns in unlabeled data, without explicit output examples.

*   **Clustering**: Grouping similar data points together (e.g., customer segmentation).
    *   *Scikit-learn examples*: `KMeans`, `DBSCAN`, `AgglomerativeClustering`.
*   **Dimensionality Reduction**: Reducing the number of features while retaining important information (e.g., for visualization or simplifying models).
    *   *Scikit-learn examples*: `PCA`, `TSNE`, `UMAP`.

### 3. Reinforcement Learning
This involves an agent learning to make decisions by interacting with an environment to maximize a reward signal. While scikit-learn doesn't directly implement RL algorithms, other libraries like `OpenAI Gym` and `Stable Baselines` are popular.

### 4. Other Important ML Functions/Steps

*   **Data Preprocessing**: Cleaning, transforming, and preparing data for models (e.g., scaling features, handling missing values, encoding categorical data).
    *   *Scikit-learn examples*: `StandardScaler`, `MinMaxScaler`, `SimpleImputer`, `OneHotEncoder`.
*   **Model Selection and Evaluation**: Choosing the best model and assessing its performance (e.g., splitting data, cross-validation, metrics).
    *   *Scikit-learn examples*: `train_test_split`, `cross_val_score`, `accuracy_score`, `mean_squared_error`.
*   **Feature Engineering**: Creating new features from existing ones to improve model performance.
*   **Hyperparameter Tuning**: Optimizing model parameters that are set before training (e.g., learning rate).
    *   *Scikit-learn examples*: `GridSearchCV`, `RandomizedSearchCV`.
